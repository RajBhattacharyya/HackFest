```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a Basemap instance
plt.figure(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=6, urcrnrlat=38, llcrnrlon=68, urcrnrlon=98, resolution='l')

# Draw map boundaries
m.drawcountries()
m.drawmapboundary(fill_color='#46bcec')
m.fillcontinents(color='#f2f2f2', lake_color='#46bcec')

# Draw states
m.drawstates()

# Coordinates for Sikkim, Kerala, and Rajasthan
state_coordinates = {
    'Sikkim': (27.5330, 88.5122),
    'Kerala': (10.8505, 76.2711),
    'Rajasthan': (27.0238, 74.2179)
}

# Plot points for Sikkim, Kerala, and Rajasthan
for state, (lat, lon) in state_coordinates.items():
    x, y = m(lon, lat)
    m.plot(x, y, 'ro', markersize=6)
    plt.text(x, y, state, fontsize=10)

plt.title('Sikkim, Kerala, and Rajasthan Highlighted on Indian Map')
plt.show()
```
