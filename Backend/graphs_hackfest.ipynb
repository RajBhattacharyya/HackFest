```python
import shapefile
import matplotlib.pyplot as plt

# Read India shapefile
india_sf = shapefile.Reader("India_Boundary.shp")

# Extract shapes and records
shapes = india_sf.shapes()
records = india_sf.records()

# Plot India map
plt.figure(figsize=(10, 8))
for shape in shapes:
    x = [i[0] for i in shape.points]
    y = [i[1] for i in shape.points]
    plt.plot(x, y, 'k')

# Define the states of interest
states_of_interest = ['Rajasthan', 'Sikkim', 'Kerala']

# Plot the states of interest
for record, shape in zip(records, shapes):
    if record[2] in states_of_interest:
        x = [i[0] for i in shape.points]
        y = [i[1] for i in shape.points]
        plt.fill(x, y, color='red', alpha=0.5)  # Adjust color and transparency as needed

# Add labels
plt.title('Rajasthan, Sikkim, and Kerala on Indian Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
